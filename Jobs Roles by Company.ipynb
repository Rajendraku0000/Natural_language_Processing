```python
df["Content"]=df.apply(lambda x : " ".join(x.dropna().astype(str)),axis=1)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer()
metrics =cv.fit_transform(df["Content"])
```

```python
similarity=cosine_similarity(metrics,metrics)
```

```python
def recommendation(Job_Role,top):
    index=df[df["Job_Role"]==Job_Role].index[0]
    new_index = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    reco=[i[0] for i in new_index[1:top+1]]
    return df.iloc[reco,:-1]
    
```

```python
recommendation("Data Scientist",10)
```

```python
recommendation("Associate Data Analyst",10)
```
